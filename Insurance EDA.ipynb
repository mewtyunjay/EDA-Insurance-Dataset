## Some questions regarding the dataset which have been solved using statistical tests


## Do charges of people who smoke differ significantly from the people who don't?

```python
df.smoker.value_counts()
```

```python
sns.stripplot(df['charges'], df['smoker'])
```

```python
# T-test to check dependency of smoking on charges

#Null Hypothesis
Ho = "Charges of smoker and non-smoker are same"   

#Alternate Hypothesis
Ha = "Charges of smoker and non-smoker are not the same"   
```

```python
#Separating charges based on smoker and non-smoker
smoker = np.array(df[df.smoker == 'yes'].charges)  
non_smoker = np.array(df[df.smoker == 'no'].charges) 
```

```python
t_statistic, p_value  = stats.ttest_ind(smoker ,non_smoker, axis = 0)

print(p_value)
```

#### Since p value is less than 0.05, we reject the null hypothesis. 
#### Hence, charges of smoker and non-smoker are not the same as 8.271435842177219e-283 < 0.05

## Does bmi of males differ significantly from that of females?

```python
df.sex.value_counts()
```

```python
sns.stripplot(df['bmi'], df['sex'])
```

```python
# T-test to check if BMI of males differ significantly from that of females

#Null Hypothesis
Ho = "BMI of males are same as that of females"   

#Alternate Hypothesis
Ha = "BMI of males are not same as that of females"   
```

```python
#Separating BMIs based on gender
men = np.array(df[df.sex == 'male'].bmi)  
women = np.array(df[df.sex == 'female'].bmi)
```

```python
t_statistic, p_value  = stats.ttest_ind(men ,women, axis = 0)

print(p_value)
```

#### P value = 0.08997637178984932 > 0.05
#### Hence, we fail to reject the null hypothesis.
#### BMI of males are same as that of females and does not differ based on gender.

## Is the proportion of smokers significantly different in different genders?


```python
#Null Hypothesis
Ho = "The proportion of smokers in different genders is same"   

#Alternate Hypothesis
Ha = "The proportion of smokers in different genders is not same"   
```

```python
#Separating smokers based on gender
female_smokers = df[df['sex'] == 'female'].smoker.value_counts()[1]  # number of female smokers
male_smokers = df[df['sex'] == 'male'].smoker.value_counts()[1] # number of male smokers
n_females = df.sex.value_counts()[1] # number of females in the data
n_males = df.sex.value_counts()[0] #number of males in the data
```

```python
print([female_smokers, male_smokers] , [n_females, n_males])
```

```python
print(f' Proportion of smokers in females, males = {round(115/662,2)}%, {round(159/676,2)}% respectively')
```

The proportions are not the same <br>
To test if they are significantly different, we will use a proportion test

```python
from statsmodels.stats.proportion import proportions_ztest

z_stat, p_value = proportions_ztest([female_smokers, male_smokers] , [n_females, n_males])

print(p_value)
```

#### Since the p value(0.005324114164320532) < 0.05, we reject the null hypothesis.
#### the proportion of smokers is significantly different in different genders

##  Is the distribution of bmi across women with no children, one child and two children, the same?

```python
# Test to check if the distribution of bmi across women with no children, one child and two children, the same

# Null Hypothesis
Ho = "The BMI are the same in the three cases"   

# Alternate Hypothesis
Ha = "The BMI are not the same"   
```

```python
#subset of original dataframe containing data of women only

female = df[df['sex'] == 'female']
```

```python
#Dividing out data into people with 0, 1 and 2 kids

zero = female[female.children == 0]['bmi']
one = female[female.children == 1]['bmi']
two = female[female.children == 2]['bmi']
```

```python
#one-way anova test

f_statistic, p_value = stats.f_oneway(zero,one,two)
```

```python
print(f"p-value is {p_value}")
```

Since p value (0.7158579926754841) > 0.05, which is our 5% acceptance level, <b>we fail to reject the null hypothesis</b>

### Therefore, the number of children, be it 0, 1 or 2 has no effect on the BMI
